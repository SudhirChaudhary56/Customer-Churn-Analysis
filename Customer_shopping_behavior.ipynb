```python
# Import libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\\Projects\\Data Analysis\\customer_shopping_behavior.csv")  
```

```python
df.head()
```

```python
df.info()
```

```python
# Summary statistics using .describe()
df.describe(include='all')
```

```python
# Checking if missing data or null values are present in the dataset
df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
# Box plot - Gender vs Purchase Amount
plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='purchase_amount', data=df)
plt.title('Gender wise spending comparison')
plt.show()

# Summary by group
print(df.groupby('gender')['purchase_amount'].mean())
```

```python
# Bar chart - Category counts
plt.figure(figsize=(10,5))
df['category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Har category mein kitne items bike')
plt.xlabel('Category')
plt.ylabel('Number of purchases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Season wise total spending
season_spend = df.groupby('season')['purchase_amount'].sum().sort_values()
season_spend.plot(kind='barh', color='lightgreen')
plt.title('Season wise total revenue')
plt.xlabel('Total Revenue (USD)')
plt.show()
```

```python
# Age distribution of customers
plt.figure(figsize=(10,5))
plt.hist(df['age'], bins=15, color='orange', edgecolor='black')
plt.title('Customer age distribution')
plt.xlabel('Age')
plt.ylabel('Number of customers')
plt.show()
```

```python
# Subscription status vs Purchase amount
sub_analysis = df.groupby('subscription_status')['purchase_amount'].agg(['mean', 'count'])
print(sub_analysis)

# Box plot
sns.boxplot(x='subscription_status', y='purchase_amount', data=df)
plt.title('Subscribers vs Non-subscribers spending')
plt.show()
```

```python
# Discount applied vs no discount - spending comparison
discount_effect = df.groupby('discount_applied')['purchase_amount'].mean()
print(discount_effect)

# Bar plot
discount_effect.plot(kind='bar', color=['red', 'green'])
plt.title('Discount effect on purchase amount')
plt.ylabel('Average purchase (USD)')
plt.show()
```

```python
# Pivot table - Category wise gender preference
pivot_table = pd.crosstab(df['category'], df['gender'], normalize='index') * 100
print(pivot_table)

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, fmt='.1f', cmap='coolwarm')
plt.title('Category preference by gender (%)')
plt.show()
```

```python
# Final Report
summary = df.groupby('category').agg({
    'purchase_amount': ['mean', 'sum', 'count'],
    'age': 'mean'
})
summary.to_csv('category_analysis.csv')

print("Analysis complete! Results saved.")
```
